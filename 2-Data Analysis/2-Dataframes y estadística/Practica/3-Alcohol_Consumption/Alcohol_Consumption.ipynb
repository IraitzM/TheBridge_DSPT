# Consumo de alcohol

Sabéis que nos preocupa mucho la salud de los ciudadanos. Por eso nos han encargado hacer un análisis sobre el consumo de alcohol.

Los datos de referencia los encontraréis aquí: https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv

Nos interesa conocer el consumo de alcohol, por tipo de bebida alcohólica para los distintos continentes y países. Además, sabiendo que el vino y la cerveza son los sospechosos habituales, podéis disponer estadísticas sobre estos dos casos en particular.

```python
import seaborn as sb
import numpy as np
import matplotlib as mpl
import pandas as pd 
```

```python
fichero_Consumo_Alcohol = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv"
user = pd.read_csv(fichero_Consumo_Alcohol, sep=',')
user
```

```python
user.set_index("continent") # SE LE ASIGNDA COMO INDICE LA VARIABLE 'continent' 
```

```python
user.describe(include="all") # SE DESPLIEGAN LOS DATOS 
```

```python
condicion = user['continent'].isnull() # SE LE ASGINAN A LA VARIABLE 'condicion' TODOS LOS DATOS NULOS DE LA COLUMNA 'continent' 

user[condicion]
```

```python
user.loc[:, 'continent'] = 'NA' # LA FUNCION .loc DEVUELVE POR PANTALLA EL RANGO DE DATOS QUE SE LE INGRESAN DE LA VARIABLE QUE SE LE INDICA 
user
```

```python
user.info() # LA FUNCION '.info()' DEVUELVE UNA ESTADISTICA GENERAL DE LA TABLA DE DATOS 
```

```python
user.groupby('continent')['beer_servings'].mean().sort_values(ascending=False) # LA FUNCION '.groupby()' REALIZA LA ACCION QUE SE LE INDICA A LOS DATOS QUE SE LE INGRESAN 
                                                                                # LOS DATOS SE INDICAN, EN PRIMER LUGAR LA VARIABLE, Y EN SEGUNDO LUGAR OTRA VARIABLE 
```

```python
grupo = user.groupby("continent")["country"].count()

print(grupo) 
```
