# Obtención de Datos Mediante APIs en Python

Alguna de las formas de obtención de datos es a través de **APIs (Application Programming Interfaces)**. Las APIs son una forma eficiente de acceder a datos y servicios en línea. Aprenderemos a realizar solicitudes a través de Python utilizando la popular biblioteca `requests`.

## Introducción a las APIs

Una API (Application Programming Interface) es un conjunto de reglas y protocolos que permiten a diferentes aplicaciones interactuar entre sí. Las APIs permiten a las aplicaciones solicitar datos o realizar acciones en un servidor remoto de manera estructurada y estandarizada.

## Protocolo HTTP

El Protocolo de Transferencia de Hipertexto (HTTP) es el protocolo subyacente utilizado en la World Wide Web. Se utiliza para solicitar y transmitir datos en la web.

* **GET**: Para recuperar datos.
* **POST**: Para enviar datos al servidor.
* **PUT**: Para actualizar datos en el servidor.

## Api chucknorris

[Enlace API](https://api.chucknorris.io/)

```python
url = "https://api.chucknorris.io/jokes/random"  # Reemplaza con la URL de la API que deseas acceder.

response = requests.get(url)

if response.status_code == 200:  # Código 200 indica una respuesta exitosa.
    data = response.json()  # Analizar la respuesta JSON.
    print(data["value"])
else:
    print("Error en la solicitud: ", response.status_code) 
```

```python
def get_joke(url): # DEVUELVE EL JASON DE UNA PETICION 
    response = requests.get(url)

    if response.status_code == 200:  # Código 200 indica una respuesta exitosa.
        data = response.json()  # Analizar la respuesta JSON.
        return data 
    else:
        print("Error en la solicitud: ", response.status_code) 
```

```python
url = "https://api.chucknorris.io/jokes/categories"
get_joke(url) # DEVUELVE CATEGORIAS 
```

Añadiendo `?` damos inicio a la secuencia de parámetros a ingresar que serán una consecución de `clave=valor` separados por `&` para cada parámetro que debamos introducir.

```python
url = "https://api.chucknorris.io/jokes/random?category=animal" # ? indica el inicio de los parámetros de consulta, category=animal es un parámetro con category como clave 
                                                                 # y animal como valor.
get_joke(url) # hace una solicitud HTTP a esa URL para obtener un chiste aleatorio de Chuck Norris en la categoría "animal".
```

# OpenStreetMap

Overpass es una instancia pública que nos permite realizar consultas sin molestar a las instancias principales de OpenStreetMap.

```python
# Se realiza una consulta a Overpass API para obtener nodos con la etiqueta "amenity=cafe" en un área específica de Madrid:

#Amenities
#https://wiki.openstreetmap.org/wiki/Key:amenity
#(43.25580545097236, -2.922715389269332, 43.25639150538353, -2.9247538815364433);

# URL de la API de Overpass
overpass_url = "http://overpass-api.de/api/interpreter" # EN LA DIRECCION DE LA API MUCHAS VECES NO SE VE NADA 

# Consulta a Overpass en formato Overpass QL
overpass_query = """
[out:json]; # ESTO ES LO QUE SE ESPERA DEN LA RESPUESTA 
node["amenity"="cafe"]
  (40.420801, -3.694702,40.423754, -3.688167); 
out;
"""

# Realiza la solicitud a la API de Overpass
response = requests.get(overpass_url, 
                        params={'data': overpass_query})

# Imprime la respuesta de la API 
print(response) 

# Convierte la respuesta a JSON
data = response.json() 
```

```python
'''El código proporciona una función json_to_dataframe que convierte datos JSON en un DataFrame de pandas.
La función toma un parámetro data_entrada, que se espera que sea un diccionario JSON. Extrae los datos bajo la
 clave 'elements' y los almacena en la variable elements. 

Se inicializa un diccionario places con listas vacías para las claves 'lat', 'lon', 'name' y 'address'. Este
diccionario se utilizará para almacenar los datos que se extraerán del JSON.

Para cada elemento en elements:

Extrae la latitud ('lat') y la longitud ('lon').
Intenta extraer el nombre ('tags'['name']). Si no existe, name será None.
Intenta extraer la calle ('tags'['addr:street']). Si no existe, se usa 'Sin calle' como valor por defecto.
Intenta extraer el número de la casa ('tags'['addr:housenumber']). Si no existe, se usa 0 como valor por
defecto.
Añade la latitud, longitud, nombre y la dirección completa (calle y número) a las listas correspondientes en
el diccionario places.

Crea un DataFrame de pandas utilizando el diccionario places y lo devuelve.

En resumen, la función toma un JSON con datos geográficos y los transforma en un DataFrame de pandas con
columnas para la latitud, longitud, nombre y dirección.

'''


def json_to_dataframe(data_entrada):        # 
    elements = data_entrada['elements']
    places = {'lat': [], 'lon': [], 'name': [], 'address': []}

    for i in elements:
        lalitude = i['lat']
        longitude = i['lon']
        name = i['tags'].get('name')
        street = i['tags'].get('addr:street', 'Sin calle')
        number = i['tags'].get('addr:housenumber', 0)
        
        places['lat'].append(lalitude)
        places['lon'].append(longitude)
        places['name'].append(name)
        places['address'].append(str(street) + ' ' + str(number))

    df = pd.DataFrame(places) 
    return df 
```

```python
'''
https://thespacedevs.com/llapi
https://ll.thespacedevs.com/2.2.0/swagger
'''

'''Este código realiza una solicitud GET a una API REST para obtener información sobre astronautas.''' 

'''este código construye una URL para la API de SpaceDevs que apunta a los datos de astronautas y luego realiza una solicitud GET a esa URL, 
almacenando la respuesta en la variable response. Para que este código funcione, es necesario tener la librería requests instalada y posiblemente 
importar dicha librería al inicio del script con import requests.'''

space_url = "http://ll.thespacedevs.com/2.2.0" # La variable contiene la url base del servicio API de SpaceDevs

path = "/astronaut" # La variable contiene el camino especifico dentro de la API para obtener información sobre astronautas  

url_total = space_url + path # Se concatenan para formar la url completa

response = requests.get(url_total) # Utiliza la librería requests para realizar una solicitud HTTP GET a la URL completa (url_total). La respuesta de esta solicitud se 
                                    # almacena en la variable response. 
```

```python
'''convierte el contenido de la respuesta HTTP, almacenada en response_sp, en un objeto de Python (normalmente un diccionario o una lista), asumiendo que el contenido es un JSON 
válido. Aquí tienes una explicación más detallada de lo que hace esta línea:

Contenido de la Respuesta:
response_sp es una variable que contiene la respuesta de una solicitud HTTP GET realizada anteriormente usando la librería requests.

Método .json():
El método .json() de un objeto Response de requests intenta analizar el contenido de la respuesta como JSON y devolver el resultado como un objeto de Python. Si el contenido de la 
respuesta es un JSON válido, este método devolverá la estructura de datos correspondiente (por ejemplo, un diccionario si el JSON representa un objeto, o una lista si el JSON 
representa un array).

Posibles Resultados:
Diccionario: Si la respuesta JSON representa un objeto, el método .json() devolverá un diccionario de Python.
Lista: Si la respuesta JSON representa un array, el método .json() devolverá una lista de Python.'''

response_sp.json() 
```

```python
'''Este código toma una respuesta de una solicitud HTTP, la cual se asume que está en formato JSON, y la convierte en un DataFrame de Pandas.
 Aquí está el desglose:

response_sp.json():
response_sp parece ser el resultado de una solicitud HTTP.
.json() es un método que generalmente se usa en bibliotecas de solicitudes HTTP para interpretar el contenido de la respuesta como JSON. 
Aquí, se espera que response_sp contenga datos en formato JSON.

['results']:
Suponiendo que el JSON devuelto por la solicitud contiene una clave denominada 'results', esta parte del código selecciona solo la parte de 
los resultados del JSON.

pd.DataFrame(...):
pd.DataFrame() es una función de Pandas que convierte diferentes tipos de datos en un DataFrame.

En este caso, se pasa el resultado de la selección de 'results' como argumento para crear un DataFrame de Pandas. Esto implica que se espera 
que 'results' contenga una lista de diccionarios, donde cada diccionario representa una fila de datos.

En resumen, este código toma los resultados de una solicitud HTTP en formato JSON, selecciona solo la parte de los resultados, y luego crea 
un DataFrame de Pandas con estos datos. Esto es útil para trabajar con datos estructurados y realizar análisis o visualizaciones en Python 
utilizando las capacidades de Pandas.  
'''

pd.DataFrame(response_sp.json()['results'])
```
